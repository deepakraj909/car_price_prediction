```python
fin_df=df3.drop(df3.columns[[0,1,2,3,4,5,6,8,12,14,15,19,20,22]],axis=1)  

fin_df
```

```python
fin_df.describe()
```

```python
def outlier_treatment(dataframe):
    Q1=dataframe.quantile(0.25)
    Q3=dataframe.quantile(0.75)
    IQR=Q3-Q1
    df_out=dataframe[~((dataframe<(Q1-1.5*IQR))|(dataframe>(Q3+1.5*IQR))).any(axis=1)]
    return df_out
                       
```

```python
final_df=outlier_treatment(fin_df)
```

```python
#outliers
plt.figure(figsize = (30,30))
plt.subplot(3,3,1)
sns.boxplot(x = 'wheel-base', y = 'price', data = final_df)
plt.subplot(3,3,2)
sns.boxplot(x = 'horsepower', y = 'price', data = final_df)
plt.subplot(3,3,3)
sns.boxplot(x = 'bore', y = 'price', data =final_df)
plt.subplot(3,3,4)
sns.boxplot(x = 'drive-wheels', y = 'price', data =final_df)
plt.subplot(3,3,5)
sns.boxplot(x = 'highway-mpg', y = 'price', data =final_df)
plt.subplot(3,3,6)
sns.boxplot(x = 'engine-size', y = 'price', data =final_df)
plt.subplot(3,3,7)
sns.boxplot(x = 'fuel-system', y = 'price', data =final_df)
```

```python
from sklearn.preprocessing import MinMaxScaler 
scaler=MinMaxScaler()
scaler.fit(final_df)
x=pd.DataFrame(scaler.transform(final_df),columns=final_df.columns)
x.head()
```

```python
#split the data into test and train
X=x.drop(columns=["price"],axis=1)
y=x["price"]
print(X)
print(y)
```

```python
#split the data imto test and train
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg_fit=reg.fit(X_train,y_train)
```

```python
reg.score(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_pred=reg_fit.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
